```python
# Regular Python
python_list_1 = list(range(1, 1000))
python_list_2 = list(range(1, 1000))

# Numpy
numpy_list_1 = np.arange(1, 1000)
numpy_list_2 = np.arange(1, 1000)

# Measure the time it takes to perform operations in regular Python
timeit_python_add = %timeit -o [(x + y) for x, y in zip(python_list_1, python_list_2)]
timeit_python_sub = %timeit -o [(x - y) for x, y in zip(python_list_1, python_list_2)]
timeit_python_mul = %timeit -o [(x * y) for x, y in zip(python_list_1, python_list_2)]
timeit_python_div = %timeit -o [(x / y) for x, y in zip(python_list_1, python_list_2)]

# Measure the time it takes to perform operations in Numpy
timeit_numpy_add = %timeit -o numpy_list_1 + numpy_list_2
timeit_numpy_sub = %timeit -o numpy_list_1 - numpy_list_2
timeit_numpy_mul = %timeit -o numpy_list_1 * numpy_list_2
timeit_numpy_div = %timeit -o numpy_list_1 / numpy_list_2

# Print the results
print("Python Addition:", timeit_python_add.best)
print("Python Subtraction:", timeit_python_sub.best)
print("Python Multiplication:", timeit_python_mul.best)
print("Python Division:", timeit_python_div.best)

print("Numpy Addition:", timeit_numpy_add.best)
print("Numpy Subtraction:", timeit_numpy_sub.best)
print("Numpy Multiplication:", timeit_numpy_mul.best)
print("Numpy Division:", timeit_numpy_div.best)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```
